# 9.Histogram of the Age Distributiion

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 10.Count Plot of Diabetes Outcome

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='diabetes', data=data)
plt.title('Count of Diabetes Outcome')
plt.xlabel('Diabetes (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

# 11. Box Plot of BMI by Diabetes Status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='bmi', data=data)
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes (1 = Yes, 0 = No)')
plt.ylabel('BMI')
plt.show()
```

# 12. Scatter Plot of Age vs Glucose level

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='blood_glucose_level', hue='diabetes', data=data)
plt.title('Age vs. Glucose Level')
plt.xlabel('Age')
plt.ylabel('Glucose Level')
plt.show()
```

# 18. Heatmap of Correlation

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Columns Only)')
plt.show()
```

# Conclusions

```python
# Diabetes Prevalence
diabetes_prevalence = data['diabetes'].mean()
print("Diabetes Prevalence:", diabetes_prevalence)
print("Conclusion: Approximately 8.82% of the individuals in the dataset have diabetes.\n")

# Grouped Statistics of Numeric Columns
numeric_data = data.select_dtypes(include=[np.number])
grouped_stats = numeric_data.groupby('diabetes').mean()
print("Grouped Statistics (Mean Values by Diabetes Status):")
print(grouped_stats)

# Conclusions from Grouped Statistics
print("\nConclusions from Grouped Statistics:")
print("1. **Age**: Individuals with diabetes are, on average, much older (60.93 years) than those without diabetes (39.94 years).")
print("2. **Hypertension**: Hypertension is 4 times more common in individuals with diabetes (24.59%) compared to those without (6.13%).")
print("3. **Heart Disease**: Heart disease is 5 times more common in individuals with diabetes (14.94%) compared to those without (3.03%).")
print("4. **BMI**: Individuals with diabetes have a much higher average BMI (32.00, obese range) compared to those without diabetes (26.87, overweight range).")
print("5. **HbA1c Level**: HbA1c levels are significantly higher in individuals with diabetes (6.93%) compared to those without (5.40%).")
print("6. **Blood Glucose Level**: Blood glucose levels are much higher in individuals with diabetes (194.03 mg/dL) compared to those without (132.82 mg/dL).\n")

# Key Insights
print("Key Insights:")
print("1. **Age is a major factor**: Diabetes is more prevalent in older individuals.")
print("2. **Obesity and BMI**: Higher BMI is strongly associated with diabetes.")
print("3. **Comorbidities**: Hypertension and heart disease are significantly more common in individuals with diabetes.")
print("4. **Diagnostic Markers**: HbA1c and blood glucose levels are significantly higher in diabetic individuals.\n")

# Recommendations
print("Recommendations:")
print("1. **Prevention**: Focus on weight management and lifestyle changes (e.g., diet, exercise) to reduce BMI and prevent diabetes, especially in older individuals.")
print("2. **Screening**: Regularly screen for hypertension and heart disease in diabetic patients, as they are at higher risk.")
print("3. **Monitoring**: Emphasize the importance of monitoring HbA1c and blood glucose levels to manage diabetes effectively.")
```
