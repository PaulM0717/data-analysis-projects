```python
#Task #9
from sklearn.model_selection import train_test_split

X = pa_df.drop(columns='target')  
y = pa_df['target']  # Target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

```python
#Task #10
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

random_state_value = 74  
dt_model = DecisionTreeClassifier(random_state=random_state_value)

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the Decision Tree model: {accuracy:.4f}")
```

```python
# Task #11
train_accuracy = dt_model.score(X_train, y_train)

test_accuracy = dt_model.score(X_test, y_test)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Testing Accuracy: {test_accuracy:.4f}")
```

```python
#Task #13
from sklearn.neighbors import KNeighborsClassifier

knn_4 = KNeighborsClassifier(n_neighbors=4)
knn_8 = KNeighborsClassifier(n_neighbors=8)

knn_4.fit(X_train, y_train)
knn_8.fit(X_train, y_train)

y_pred_4 = knn_4.predict(X_test)
y_pred_8 = knn_8.predict(X_test)

accuracy_4 = accuracy_score(y_test, y_pred_4)
accuracy_8 = accuracy_score(y_test, y_pred_8)

print(f"Accuracy with K=4: {accuracy_4:.4f}")
print(f"Accuracy with K=8: {accuracy_8:.4f}")
```

```python
#Task #14
train_accuracy_4 = knn_4.score(X_train, y_train)
test_accuracy_4 = knn_4.score(X_test, y_test)

train_accuracy_8 = knn_8.score(X_train, y_train)
test_accuracy_8 = knn_8.score(X_test, y_test)

print(f"Training Accuracy with K=4: {train_accuracy_4:.4f}")
print(f"Testing Accuracy with K=4: {test_accuracy_4:.4f}")

print(f"Training Accuracy with K=8: {train_accuracy_8:.4f}")
print(f"Testing Accuracy with K=8: {test_accuracy_8:.4f}")
```

```python
#Task #15
better_knn = knn_4 if test_accuracy_4 > test_accuracy_8 else knn_8
y_pred_better_knn = better_knn.predict(X_test)

cm_better_knn = confusion_matrix(y_test, y_pred_better_knn)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_better_knn, annot=True, fmt='d', cmap='Blues', xticklabels=wine.target_names, yticklabels=wine.target_names)
plt.title(f"Confusion Matrix for KNN (K={better_knn.n_neighbors})")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
